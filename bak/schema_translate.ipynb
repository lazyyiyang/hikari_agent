```python
import akshare as ak
import pandas as pd
from langchain_openai.chat_models import ChatOpenAI
from dotenv import load_dotenv
load_dotenv(override=True)  # 确保这行在ChatOpenAI初始化之前执行
from langchain_core.prompts import ChatPromptTemplate,PromptTemplate
import ast
import os
```

```python
symbol= 'hk'
```

```python
llm = ChatOpenAI(
    api_key=os.getenv("API_KEY", ""),  # type: ignore
    base_url="https://api.deepseek.com",
    model="deepseek-chat",
)
```

```python
system_prompt = """您是一位资深金融分析师，需要根据以下要求进行分析：
1. 严格使用提供的财务数据
2. 保持专业严谨的分析风格
3. 所有结论必须有数据支撑"""
```

```python
# 定义包含参数占位符的人类提示模板
human_template = "{input}"
```

```python
def df_scehma_translate(df,df_info,system_prompt,query):
    # 创建包含系统提示词的模板
    
    prompt = ChatPromptTemplate.from_messages([
    ("system", system_prompt),
    ("human",  query) 
    ])
    # 将提示词模板与模型组合成链
    chain = prompt | llm
    response = chain.invoke({
    "input": query
    })
    response_res = (response.content)
    #print (response_res)
    #response_res = response_res.split("\n\n")[1]
    result_list = ast.literal_eval(response_res) 
    result_dict = dict(zip(result_list, df_info))
    return result_dict
```

```python
#query1 = "资产负债表"
df1 = ak.stock_balance_sheet_by_report_em(symbol=symbol)
```

```python
df1_list = list(df1.columns)
query1 = f"请根据资产负债表对以下英文列表进行一对一翻译{df1_list}，结果要输出列表list"
bal_result_dict = df_scehma_translate(df1,df1_list,system_prompt,query1)
```

```python
df2 = ak.stock_profit_sheet_by_report_em(symbol=symbol)
df2_list = list(df2.columns)
query2 = f"请根据利润表对以下英文列表进行一对一翻译{df2_list}，结果要输出列表list"
profit_result_dict = df_scehma_translate(df2,df2_list,system_prompt,query2)
```

```python
df3 = ak.stock_cash_flow_sheet_by_report_em(symbol=symbol)
df3_list = list(df3.columns)
query3 = f"请根据现金流量表对以下英文列表进行一对一翻译{df3_list}，结果要输出列表list,不要输出其他补充解释"
cf_result_dict = df_scehma_translate(df3,df3_list,system_prompt,query3)
```
